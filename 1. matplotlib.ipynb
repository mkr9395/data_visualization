### Plot Horizontal and vertical lines on Graph :

```python
# scatter plot on avg and strike rate using plt.scatter()
plt.figure(figsize=(15,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.xlabel('average runs')
plt.ylabel('strike rate')

plt.axhline(130,color='red',label='strike rate = 130')
plt.axhline(140,color='blue',label='strike rate = 140')
plt.axvline(30,color='green',label='average = 30')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'][i],sample_df['strike_rate'][i],sample_df['batter'][i])
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1, sharex=True, figsize=(10,6)) # share x-axis and figure size
 
ax[0].scatter(batters['avg'], batters['strike_rate'], color = 'red')
ax[1].scatter(batters['avg'], batters['runs'])

ax[0].set_title('Avg vs Strike rate')
ax[1].set_title('Avg vs total runs')

ax[0].set_ylabel('Strike rate')

ax[1].set_xlabel('Avg')
ax[1].set_ylabel('total runs')

fig.show()
```

##### creating subplots : 

```python
fig, ax = plt.subplots(nrows=2,ncols=2, figsize=(10,10)) # share x-axis and figure size
 
ax[0,0].scatter(batters['avg'], batters['strike_rate'], color = 'red')
ax[0,1].scatter(batters['avg'], batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])

ax[0,0].set_title('Avg vs Strike rate')
ax[0,1].set_title('Avg vs total runs')
ax[1,0].set_title('Hist for Avg')
ax[1,1].set_title('Hist for Runs')

ax[0,0].set_ylabel('Strike rate')
ax[0,1].set_ylabel('total runs')
ax[1,0].set_ylabel('number of players')
ax[1,1].set_ylabel('number of players')


# ax[1].set_xlabel('Avg')
# ax[1].set_ylabel('total runs')

fig.show()
```

## 3D Surface Plots:

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
```

```python
xx
```

```python
yy
```

```python
xx.shape, yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Countour Plots : 3-D plots in 2-D

```python
fig = plt.figure(figsize=(10,6))
ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis') #countrf
fig.colorbar(p)
```
